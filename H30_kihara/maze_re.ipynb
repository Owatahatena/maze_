```python
import random
import itertools


def gen_maze(maze_size):

    W = maze_size

    maze = [[0 if i == 0 or j == 0 or j == W-1 or i == W-1 else 1 for j in range(W)] for i in range(W)] #通路:0 壁:1

#for ma in maze:
#    print(ma)

    l = range(2,W-1)
#print(list(itertools.combinations(list(l)[::2],2)))
    search_targets = list(itertools.product(list(l)[::2],list(l)[::2]))
#print(search_targets)

    target = search_targets.pop(random.randint(0, len(search_targets)-1))

#print(target)

    end_search = [target]

    maze[target[1]][target[0]] = 0

    offset = [[-1,0],[1,0],[0,-1],[0,1]]


    while(len(search_targets) > 0):
        random.shuffle(offset)
        for i in range(4):
            search_x = target[0]+offset[i][0]*2
            search_y = target[1]+offset[i][1]*2
            if maze[search_y][search_x] == 1:
            #print((search_x,search_y))
            #print(search_targets)
                search_targets.remove((search_x,search_y))
                end_search.append((search_x,search_y))
                maze[search_y][search_x] = 0
                maze[search_y-offset[i][1]][search_x-offset[i][0]] = 0
                break
                if i == 3:
                    end_search.remove((search_x,search_y))
            target = end_search[random.randint(0,len(end_search)-1)]
            
            
    return maze

            



# for m in maze:
#     print(m)


```

```python
class State:
    def __init__(self):
        self.a = 1

class Gohome(State):
    def __init__(self,ant):
        self.ant = ant
        self.direction =[1,2,4,4]
        
    def change_state(self):
        self.ant.state = self.ant.search() 

class Search(State):
    def __init__(self,ant):
        self.ant = ant
        
    def _around(self,offset,Map):
        now = (ant.coordinate[0],ant.coordinate[1])
        for ofs in offset:
            ofs_now = np.array(now) + ofs
            #print('ofs_now',ofs_now)
            #print('now',now)
            
            if Map[ofs_now[0]][ofs_now[1]] == 0:
                yield ofs
        
    def search(self,offset,Map):
        offset = self._around(offset,Map)
        offset = list(offset)
        r = random.choice(offset)
        self.ant.coordinate = self.ant.coordinate + np.array(r) 
        return r
        
    def change_state(self):
        self.ant.state = self.ant.gohome()

class Ant():
    def __init__(self):
        self.gohome_state = Gohome(self)
        self.search_state = Search(self)
        self.state = self.search_state
        
        self.coordinate = np.array((2,2))
        
    
    def change_state(self):
        self.state = self.state.change_state()
        
```
